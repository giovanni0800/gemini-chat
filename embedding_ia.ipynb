```python
# Listagem de documentos que serão buscados
DOCUMENT1 = {
    "Título": "O significado de Estatística",
    "Conteúdo": "A estatística investiga, relaciona e apresenta dados sobre populações e fenômenos de diferentes naturezas. A estatística investiga, analisa e representa dados de uma pesquisa. Estatística é a ciência que estuda a coleta, estrutura, interpretação e representação de informações."
}

DOCUMENT2 = {
    "Título": "O significado de KPI",
    "Conteúdo": "É um indicador que vai nos dar um contexto de meta (objetivo). Ex.:  Esse mês vendemos 200% do nosso objetivo."
}

DOCUMENT3 = {
    "Título": "O significado de OKR",
    "Conteúdo": "É um framework (ou processo de gestão) que vai utilizar das KPIs escolhidas para criar um direcionamento claro dos objetivos da empresa como um todo e ajudar a acompanhar o atingimento desses objetivos."
}

DOCUMENT4 = {
    "Título": "Explicando o futebol",
    "Conteúdo": "O futebol é o esporte mais popular do mundo, com mais de 250 milhões de jogadores ativos em mais de 200 países. É praticado por pessoas de todas as idades, sexos e origens sociais, e é considerado um símbolo de união, trabalho em equipe e fair play."
}

documents = [DOCUMENT1, DOCUMENT2, DOCUMENT3, DOCUMENT4]
print("Ok!")
```

```python
df = pd.DataFrame(documents)
df.columns = ["Titulo", "Conteudo"]

display(df)
```

```python
model = "models/embedding-001"
print("Ok!")
```

```python
def embed_fn(title, text):
    return genai.embed_content(model=model, title=title, content=text, task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

```python
# Aqui eu estou criando uma coluna nova chamada Embeddings,
# estou percorrendo cada linha dentro do Data Frame e adicionando os valores de embedding da função embed_fn a cada uma dessas
# linhas dentro da nova coluna Embeddings.
df["Embeddings"] = df.apply( lambda row: embed_fn( row["Titulo"], row["Conteudo"] ), axis=1 ) # O axis=1 indica que quero aplicar essa função de embed nas linhas do Embeddings (já que o apply vai percorrer todas as linhas ou todas as colunas)

display(df)
```

```python
# Aqui eu crio uma correlação entre a consulta e o embedding retornado por ela e os embeddings das respostas armazenadas no Data Frame.
# Gero um índice da localização onde os embeds que foram gerados pela pergunta batem com os embeds do DF,
# assim retornando um índice, o qual uso para buscar o índice da coluna do Conteúdo para retornar o texto da linha condizente
def gerar_buscar_consulta(consulta, base, modelo):
    embedding_da_consulta = genai.embed_content(model=modelo,
                                                content=consulta,
                                                task_type="RETRIEVAL_QUERY")["embedding"]
    produtos_escalares = np.dot( np.stack( df["Embeddings"] ), embedding_da_consulta )

    indice = np.argmax( produtos_escalares )
    return df.iloc[indice]["Conteudo"]
```

```python
consulta = "O que é Futebol?"

trecho = gerar_buscar_consulta(consulta=consulta, base=df, modelo=model)

print(trecho)
```

```python
generation_config = {
    "temperature": 0.5,
    "candidate_count": 1,
}
```

```python
prompt = f"Reescreva, de uma forma mais descontraída e compreensível, sem tirar ou adicionar informações, o trecho: {trecho}"

model_gemini = genai.GenerativeModel(model_name="gemini-1.0-pro",
                                     generation_config=generation_config)
response = model_gemini.generate_content(prompt)

print( response.text )
```
